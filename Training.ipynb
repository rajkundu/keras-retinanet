```python
import os
import urllib

# Resume training from epoch
RESUME_EPOCH = 142
# Total num epochs to train (including already completed)
TOTAL_NUM_EPOCHS = 200

PRETRAINED_MODEL_PATH = './snapshots/resnet50_csv_' + str(RESUME_EPOCH) + '.h5'
if(os.path.exists(PRETRAINED_MODEL_PATH)):
    print('Using existing pretrained model at ' + PRETRAINED_MODEL_PATH)
else:
    URL_MODEL = 'https://github.com/fizyr/keras-retinanet/releases/download/0.5.1/resnet50_coco_best_v2.1.0.h5'
    urllib.request.urlretrieve(URL_MODEL, PRETRAINED_MODEL_PATH)
    print('Downloaded pretrained model to ' + PRETRAINED_MODEL_PATH)
```

```python
# Run training python script
!keras_retinanet/bin/train.py \
--freeze-backbone \
--random-transform \
--weights {PRETRAINED_MODEL_PATH} \
--batch-size 1 \
--steps 500 \
--epochs {TOTAL_NUM_EPOCHS} \
--initial-epoch {RESUME_EPOCH} \
--debug-gpu 0 \
csv annotations.csv classmap.csv
```
